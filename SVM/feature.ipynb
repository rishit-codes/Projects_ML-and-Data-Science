```python
# Mask correlations below threshold
corr = df.corr(numeric_only=True)

mask = abs(corr) < 0.75
corr_masked = corr.mask(mask)

plt.figure(figsize=(14, 12))
sns.heatmap(corr_masked, annot=True, cmap='Oranges')
```

```python
# this function is used to remove the columns whose correlation matrix has a value
# exceeding than the threshold

def corre(dataset,threshold):

    numeric_data = dataset.select_dtypes(include=['number'])

    col_corr = set()
    corr_matrix = numeric_data.corr()
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j] > threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

```python
y.info()
```

```python
print(X.shape)   
print(y.shape)  
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.30,random_state=42)
scaler = StandardScaler()
```

```python
corr_features = corre(X_train,0.75)

X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
print(X_train.shape,X_test.shape)

X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

svc = SVC(kernel='linear')                  ## LINEAR

svc.fit(X_train_scaler,y_train)
y_pred = svc.predict(X_test_scaler)


print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
## rbf
svc = SVC(kernel='rbf')

svc.fit(X_train_scaler,y_train)
y_pred = svc.predict(X_test_scaler)


print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))

```

```python
## poly
svc = SVC(kernel='poly')

svc.fit(X_train_scaler,y_train)
y_pred = svc.predict(X_test_scaler)


print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# sigmoid
svc = SVC(kernel='sigmoid')

svc.fit(X_train_scaler,y_train)
y_pred = svc.predict(X_test_scaler)


print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

HYPERPARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 'C':[0.01,0.1,1,10,100,1000],
              'gamma':[1,0.1,0.01,0.0001,0.001],
              'kernel':['linear','rbf','sigmoid','poly']
}

grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, refit=True, verbose=3, cv=5)
```

```python
grid.fit(X_train_scaler,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test_scaler)

print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
